```python
#Using UCB
import random
import math

def ucb(restaurants, budget):
    restaurant_qualities = [random.uniform(0, 1) for _ in range(restaurants)]  # Simulated restaurant qualities
    visits = [0] * restaurants  # Number of times each restaurant has been visited
    total_quality = [0] * restaurants  # Total quality collected for each restaurant

    for t in range(1, budget + 1):
        ucb_values = [total_quality[i] / max(1, visits[i]) + math.sqrt(2 * math.log(t) / max(1, visits[i])) for i in range(restaurants)]
        selected_restaurant = ucb_values.index(max(ucb_values))
        visits[selected_restaurant] += 1

        # Simulate visiting the selected restaurant and get its actual quality rating
        actual_quality = random.uniform(0, 1)  # Assuming ratings range from 0 to 12
        total_quality[selected_restaurant] += actual_quality

        print(f"Visited Restaurant {selected_restaurant + 1} (Quality: {actual_quality:.2f})")

    recommended_restaurant = visits.index(max(visits))
    print(f"Recommended Restaurant to Visit: Restaurant {recommended_restaurant + 1} (Quality: {restaurant_qualities[recommended_restaurant]:.2f})")

if __name__ == "__main__":
    restaurants = int(input("Enter the number of restaurants: "))
    budget = int(input("Enter the budget for restaurant visits: "))

    ucb(restaurants, budget)
```
